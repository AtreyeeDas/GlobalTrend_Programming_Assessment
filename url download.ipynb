```python
import requests
import time
```

```python
def download_urls_with_retries(urls, max_retries=3, sleep_time=2):
  results = []
  for url in urls:
    for attempt in range(max_retries):
      try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for error HTTP status codes
        results.append(response.content)
        break  # Successful download, break out of retry loop
      except requests.exceptions.RequestException as e:
        print(f"Error downloading {url}: {e}")
        if attempt == max_retries - 1:
          print(f"Max retries reached for {url}")
          results.append(None)  # Indicate failure
        else:
          time.sleep(sleep_time)  # Wait before retrying
  return results
```
