```python
import time
import logging

def execution_time_logger(func):
  def wrapper(*args, **kwargs):
    start_time = time.time()
    result = func(*args, **kwargs)
    end_time = time.time()
    execution_time = end_time - start_time
    logger = logging.getLogger(__name__)
    logger.info(f"Function {func.__name__} executed in {execution_time:.4f} seconds")
    return result
  return wrapper
```

```python
# Example usage:
def computationally_expensive_function():
  # Simulate a computationally expensive task
  time.sleep(2)
  return "Task completed"
```

```python
# Apply the decorator
computationally_expensive_function = execution_time_logger(computationally_expensive_function)
```

```python
# Call the function
result = computationally_expensive_function()
print(result)
```
